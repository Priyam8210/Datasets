# handling missing data in column 'imdb_id'

```python
for idx, con in enumerate(df_tb.imdb_id):
  if pd.isnull(con):
    df_tb.imdb_id[idx] = df_tb.new_imdb[idx]
```

```python
def ext_feat(data, threshold):
  cor_ini = set()
  for i in range(len(data.columns)):
    for j in range(i):
      if abs(data.iloc[i, j]) > threshold:
        da = data.columns[i]
        cor_ini.add(da)

  return cor_ini

```

```python
corelated_feature = ext_feat(cor_map, .8)
corelated_feature
```
